```python
#1번
def fibo_loop(n):
  value1=0
  value2=1
  for i in range(1,n):
    if( i==1):
      result=0
      print (result)
    elif(i==2):
      result=1
      print (result)
    else:
      result=value1+value2
      value1=value2
      value2=result
      print (result)

def fibo_recur(n):
  if n==1:
    return 0
  elif n==2:
    return 1
  else:
    return fibo_recur(n-1)+fibo_recur(n-2)
#2번
def Hanoi(n , now, des, other):#현재 위치 now, 목적지 des, 보조 기둥 other 
    global check#원반을 움직이는 총 횟수를 나타내는 check변수
    check+=1
    if n==1:
        print ("Move disk 1 from",now,"to",des)
    else:
      Hanoi(n-1, now, other, des)
      print ("Move disk",n,"from",now,"to",des)
      Hanoi(n-1, other, des, now)
        
#1번
fibo_loop(41)
#fibo_recur(41)

#2번
n = 10
check=0
Hanoi(n,'LEFT','RIGHT','CENTER')
print("total number of times:",check)
# LEFT, CENTER, RIGHT are the name of rods
```
